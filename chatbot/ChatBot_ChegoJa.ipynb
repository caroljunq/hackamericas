Um algoritmo de Machine Learning é usado pra treinar o bot a partir de uma lista no padrão pergunta-resposta. O modelo permite uma certa flexibilidade na digitação do input. Aqui trata-se do mesmo item, pois o bot confirma a escolha.  
Apesar da relativa lentidão da implementação, que leva alguns minutos para treinar o modelo, não trata-se de um problema: esse comando é rodado apenas uma vez no back end, e o modelo fica disponível para interação.  
Na interação por WhatsApp alguns segundos entre pergunta e resposta é considerado normal e até mesmo confere certa naturalidade a linguagem do bot.  

```python
fp = 'lista_de_linhas.txt'
qual_linha = ChatBot('qual_linha')

qual_linha.set_trainer(ListTrainer)

with open(fp) as onibus_file:
    for line in onibus_file:
        
    # o código é divido em ida e volta para registrar o sentido
        if line != '\n':
            line = line.split('-')
            
            onibus_lista = []
            ida = '{0} - {1}'.format(line[0].strip(), line[1].strip())
            for i in range(2):
                onibus_lista.append(ida)
            qual_linha.train(onibus_lista)
    
            
            onibus_lista = []
            volta = '{0} - {1}'.format(line[1].strip(), line[0].strip())
            for i in range(2):
                onibus_lista.append(volta)
            qual_linha.train(onibus_lista)

qual_linha.read_only = True
```

```python
linhas_cod = []
linhas_string = []
linhas_entities = {}

string_linhas_fp = 'lista_de_linhas.txt'
cod_linhas_fp = 'cod_de_linhas.txt'


with open(cod_linhas_fp) as file:
    for line in file:
        cod_linha = (line[:-2])
        
        if line != '\n':
            linhas_cod.append(cod_linha)
        

with open(string_linhas_fp) as onibus_file:
    for line in onibus_file:
        if line != '\n':
            linha = []
            line = line.split('-')
            
            ida = '{0} - {1}'.format(line[0].strip(), line[1].strip())
            linha.append(ida)
            
            volta = '{0} - {1}'.format(line[1].strip(), line[0].strip())
            linha.append(volta)
            
            linhas_string.append(linha)
            
dict_linhas = dict(zip(linhas_cod, linhas_string))



for k in dict_linhas.keys():
    linhas_entities.setdefault(k)
    linhas_entities[k] = Linha(k)
    

print(dict_linhas)
```
