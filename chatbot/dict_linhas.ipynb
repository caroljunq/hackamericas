```python
cod_linhas_fp = 'cod_de_linhas.txt'

linhas_cod = []
with open(cod_linhas_fp) as file:
    for line in file:
        cod_linha = (line[:-2])
        
        if line != '\n':
            linhas_cod.append(cod_linha)
        

print(linhas_cod)
```

```python
string_linhas_fp = 'lista_de_linhas.txt'

linhas_string = []
with open(string_linhas_fp) as onibus_file:
    for line in onibus_file:
        if line != '\n':
            linha = []
            line = line.split('-')
            
            ida = '{0} - {1}'.format(line[0].strip(), line[1].strip())
            linha.append(ida)
            
            volta = '{0} - {1}'.format(line[1].strip(), line[0].strip())
            linha.append(volta)
            
            linhas_string.append(linha)         
```

```python
dict_linhas = dict(zip(linhas_cod, linhas_string))
```

```python
class Linha:
    
    def __init__(self, cod):
        self.users = []
        self.cod = cod
        self.ida = dict_linhas[cod][0]
        self.volta = dict_linhas[cod][1]

    def add_user(self, user_id):
        if user_id not in self.users:
            self.users.append(user_id)
        


linhas_entities = {}

for k in dict_linhas.keys():
    linhas_entities.setdefault(k)
    linhas_entities[k] = Linha(k)
```

```python
print(linhas_entities['118Y-1'].cod)
print(linhas_entities['118Y-1'].ida)
print(linhas_entities['118Y-1'].volta)

linhas_entities['118Y-1'].add_user('Felipe_dos_Anjos')
linhas_entities['118Y-1'].add_user('Carol_Junqueira')

print(linhas_entities['118Y-1'].users)
```
